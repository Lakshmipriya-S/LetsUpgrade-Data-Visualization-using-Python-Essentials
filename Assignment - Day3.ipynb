### Import libraries

```python
import seaborn as sns
sns.set_style('darkgrid')
```

### Loading dataset from seaborn library

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
fmri.subject.nunique()
```

### Visualisation

```python
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=2)
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=2,palette="OrRd")
```

### Line plot

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=2,kind='line')
```

```python

sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=2,kind='line',palette="Dark2_r")
```

### Boxplot

```python

sns.boxplot(x='timepoint',y='signal',hue='event',data=fmri,palette="PuRd_r",fliersize=4)
```

```python
sns.boxplot(x='timepoint',y='signal',hue='region',data=fmri,palette="Dark2_r")
```
