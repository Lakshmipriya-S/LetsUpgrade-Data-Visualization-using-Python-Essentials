## Question 1

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

### Line plot

```python
plt.plot(a,b,'c--v',linewidth=2,markersize=12,markerfacecolor='b')
plt.xlabel('value range from 40 to 50')
plt.ylabel('value range from 50 to 60')
plt.title('Line plot')
```

## Question 2

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

### Line plot for sales_1 and sales_2, for all days in a week

```python
plt.plot(days,sales_1,'red',label='Sales_1')
plt.plot(days,sales_2,'black',label='Sales_2')
plt.xlabel('Days',color='c')
plt.ylabel('Sales',color='c')
plt.title('Days Vs Sales',color='c')
plt.legend(loc='lower right')
plt.grid()
```

## Question 3

### create subplots

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

### Creating subplots

```python

plt.subplot(3,3,1)
plt.plot(x,y1,'r')
plt.subplot(3,3,2)
plt.plot(x,y2,'b')
plt.subplot(3,3,3)
plt.plot(x,y3,'y')
plt.subplot(3,3,4)
plt.plot(x,y4,'orange')
plt.subplot(3,3,5)
plt.plot(x,y5,'c')
```

### creating subplots based on axis index

```python
fig,axs=plt.subplots(2,3,figsize=(6,4))
axs[0,0].plot(x,y1)
axs[0,1].plot(x,y2)
axs[0,2].plot(x,y3)
axs[1,0].plot(x,y4)
axs[1,1].plot(x,y5)
```
